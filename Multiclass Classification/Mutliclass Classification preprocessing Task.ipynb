# Preprocessing step for bending1

```python
try :
    bending1_df = pd.DataFrame(columns = ["time","avg_rss12","var_rss12","avg_rss13","var_rss13","avg_rss23","var_rss23", "label"])
    for i in os.listdir("bending1"):    
        f = open(f"bending1//{i}", "r").readlines()
        data  = []
        
        for i in f[5:]:
            l1 = str([i]).split(",")
            data.append(l1)
            
        df = pd.DataFrame(data)  
        
        for i in range(len(df)):
            df[0][i] = int(df[0][i].split("['")[1])
        
        for i in range(len(df)):
            df[6][i] = float(df[6][i].split("\\n")[0])
        
        df["label"] = "bending1"
        col_names =  ["time","avg_rss12","var_rss12","avg_rss13","var_rss13","avg_rss23","var_rss23", "label"]
    
        df.columns = col_names
        df["avg_rss12"] = pd.to_numeric(df["avg_rss12"])
        df["var_rss12"] = pd.to_numeric(df["var_rss12"])
        df["avg_rss13"] = pd.to_numeric(df["avg_rss13"])
        df["var_rss13"] = pd.to_numeric(df["var_rss13"])
        df["avg_rss23"] = pd.to_numeric(df["avg_rss23"])
        df["var_rss23"] = pd.to_numeric(df["var_rss23"])
    
        bending1_df = pd.concat([df, bending1_df])
        
    bending1_df = bending1_df.reset_index() # resetting the index 
    bending1_df.drop("index", axis =1, inplace = True) # dropping the index column
 
except Exception as e :
    print("Error occured ", e)
```
